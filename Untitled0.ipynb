```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Data
x = np.array([1, 2, 3, 4, 5, 6]).reshape(-1, 1)   # Load (N)
y = np.array([30, 38, 24, 21, 78, 45])            # Deformation (mm)

# Polynomial regression (2nd degree)
poly = PolynomialFeatures(degree=2)
x_poly = poly.fit_transform(x)

model = LinearRegression()
model.fit(x_poly, y)

# Smooth curve
x_smooth = np.linspace(1, 6, 200).reshape(-1, 1)
x_smooth_poly = poly.transform(x_smooth)
y_smooth = model.predict(x_smooth_poly)

# Plot
plt.figure(figsize=(8,5))
plt.scatter(x, y, s=80, label='Experimental Data')
plt.plot(x_smooth, y_smooth, linewidth=2, label='Regression Curve')

# Labels and title
plt.xlabel('Load (N)', fontsize=12)
plt.ylabel('Deformation (mm)', fontsize=12)
plt.title('Load vs Deformation Regression Curve', fontsize=14)

plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Cursor
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# -----------------------------
# Experimental Data
# -----------------------------
x = np.array([1, 2, 3, 4, 5, 6])          # Load (N)
y = np.array([30, 38, 24, 21, 78, 45])    # Deformation (mm)

# -----------------------------
# Polynomial Regression
# -----------------------------
degree = 2

poly = PolynomialFeatures(degree=degree)
x_poly = poly.fit_transform(x.reshape(-1, 1))

model = LinearRegression()
model.fit(x_poly, y)

# Smooth curve generation
x_smooth = np.linspace(min(x), max(x), 300)
x_smooth_poly = poly.transform(x_smooth.reshape(-1, 1))
y_smooth = model.predict(x_smooth_poly)

# Regression Equation
coef = model.coef_
intercept = model.intercept_

equation = (
    f"y = {coef[2]:.4f}x² "
    f"{coef[1]:+.4f}x "
    f"{intercept:+.4f}"
)

# -----------------------------
# Plotting
# -----------------------------
fig, ax = plt.subplots(figsize=(10, 6))

# Scatter points
scatter = ax.scatter(
    x,
    y,
    s=100,
    marker='o',
    label='Experimental Data'
)

# Regression curve
ax.plot(
    x_smooth,
    y_smooth,
    linewidth=2.5,
    label='Polynomial Regression Curve'
)

# Labels and title
ax.set_title('Load vs Deformation Regression Curve', fontsize=16)
ax.set_xlabel('Load (N)', fontsize=13)
ax.set_ylabel('Deformation (mm)', fontsize=13)

# Grid
ax.grid(True, linestyle='--', alpha=0.6)

# Equation text
ax.text(
    0.05,
    0.92,
    equation,
    transform=ax.transAxes,
    fontsize=12,
    bbox=dict(facecolor='white', alpha=0.8)
)

# Cursor interaction
cursor = Cursor(ax, useblit=True, linewidth=1)

# -----------------------------
# Hover Annotation
# -----------------------------
annot = ax.annotate(
    "",
    xy=(0, 0),
    xytext=(15, 15),
    textcoords="offset points",
    bbox=dict(boxstyle="round", fc="w"),
    arrowprops=dict(arrowstyle="->")
)

annot.set_visible(False)

def update_annot(ind):
    pos = scatter.get_offsets()[ind["ind"][0]]
    annot.xy = pos
    text = (
        f"Load = {pos[0]:.0f} N\n"
        f"Deformation = {pos[1]:.0f} mm"
    )
    annot.set_text(text)

def hover(event):
    vis = annot.get_visible()

    if event.inaxes == ax:
        cont, ind = scatter.contains(event)

        if cont:
            update_annot(ind)
            annot.set_visible(True)
            fig.canvas.draw_idle()
        else:
            if vis:
                annot.set_visible(False)
                fig.canvas.draw_idle()

fig.canvas.mpl_connect("motion_notify_event", hover)

# Legend
ax.legend()

# Show plot
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

# -----------------------------
# Dataset
# -----------------------------
X = np.array([1, 2, 3, 4, 5, 6]).reshape(-1, 1)
y = np.array([30, 38, 24, 21, 78, 45])

# -----------------------------
# Models
# -----------------------------
models = {
    "Linear Regression": LinearRegression(),

    "Polynomial Regression":
        make_pipeline(
            PolynomialFeatures(degree=2),
            LinearRegression()
        ),

    "Decision Tree":
        DecisionTreeRegressor(max_depth=3, random_state=42),

    "Random Forest":
        RandomForestRegressor(
            n_estimators=100,
            random_state=42
        )
}

# -----------------------------
# Train & Evaluate
# -----------------------------
results = []

plt.figure(figsize=(10,6))

# Original data
plt.scatter(X, y, s=100, label='Actual Data')

x_smooth = np.linspace(1, 6, 300).reshape(-1,1)

for name, model in models.items():

    model.fit(X, y)

    y_pred = model.predict(X)

    # Metrics
    r2 = r2_score(y, y_pred)
    mae = mean_absolute_error(y, y_pred)
    rmse = np.sqrt(mean_squared_error(y, y_pred))

    results.append([name, r2, mae, rmse])

    # Smooth prediction curve
    y_smooth = model.predict(x_smooth)

    plt.plot(x_smooth, y_smooth, linewidth=2, label=name)

# -----------------------------
# Result Table
# -----------------------------
results_df = pd.DataFrame(
    results,
    columns=["Model", "R² Score", "MAE", "RMSE"]
)

print("\nModel Performance:\n")
print(results_df)

# -----------------------------
# Plot
# -----------------------------
plt.title("ML Regression Models Comparison", fontsize=16)
plt.xlabel("Load (N)", fontsize=13)
plt.ylabel("Deformation (mm)", fontsize=13)

plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()

plt.tight_layout()
plt.show()
```
